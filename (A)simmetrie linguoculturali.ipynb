In linguistica applicata si può considerare una lingua naturale come *linguocultura*, ossia come l'unione inscindibile di due ordini di elementi. Il primo è un ordine sociale, l'ordine del *ritaglio*, composto dai fenomeni di natura culturale secondo cui una comunità di parlanti ritaglia la realtà in concetti e li assegna a segni linguistici. Il secondo è un ordine fisico, l'ordine della *tessitura*, composto dai meccanismi secondlo cui i fili dei segni linguistici vengono intrecciati per comporre un testo.

Più piccola la comunità dei parlanti, più specifici saranno i concetti che è necessario veicolare. Questa specificità può essere legata sia alla limitata diffusione dei concetti, sia alla limitata diffusione dei segni linguistici. Il risultato è un tessuto che può sembrare inestricabile, incomprensibile a chi non appartiene alla specifica comunità linguistica. La *terminologia* si occupa di studiare questi fenomeni quando i concetti da veicolare si possono ricondurre a un sapere di tipo specialistico o settoriale.

![Scissors and thread](http://media-cache-ak0.pinimg.com/736x/27/93/71/279371e942e9f5a7d4224901e747710c.jpg)

Il legame linguoculturale tra sistema fisico di segni linguistici e sistema di concetti condivisi è particolarmente radicato in campo giuridico. Ogni linguocultura costruisce e si caratterizza per un diritto che le è peculiare e irripetibile, in quanto è il prodotto di un’evoluzione nel corso della storia. La forte componente interpretativa che entra in gioco nella terminologia giuridica rende difficile, se non pericolosa, la trattazione della parte concettuale senza una conoscenza approfondita delle complesse dinamiche storico sociali in questione. Per questa ragione proseguiremo questa analisi soffermandoci sulle dinamiche dei segni linguistici.
